# Import Package

```python
import numpy as np
import pandas as pd
```

# Menyimpan dataset kedalam variabel "data"

```python
data = pd.read_csv('cwurData.csv')
```

# Menampilkan record dataset

```python
data
```

# Menampilkan informasi jumlah record, nama attribute, jumlah attribute dengan tipe objek, jumlah attribute dengan tipe angka, dan jumlah attribute yang memiliki nilai null

```python
jumlahRecord = len(data)
atribut = data.columns
objek = len(data.select_dtypes(include=['object']).columns)
float = len(data.select_dtypes(include=['float64']).columns)
angka = len(data.select_dtypes(include=['int64']).columns)
null = len(data.columns[data.isna().sum() > 0])

dictionary = {
    "Jumlah Record = ": jumlahRecord,
    "Nama Atribut  = ": atribut, 
    "Jumlah Objek  = ": objek,
    "Jumlah Float  = ": float,
    "Jumlah Angka  = ": angka, 
    "Jumlah Null   = ": null,
}

pd.Series(dictionary)
```
