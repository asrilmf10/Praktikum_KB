Normalisasi pada salah satu atribut (Min Max scaler)

```python
MMScaler = MinMaxScaler()

data_normalisasi = MMScaler.fit_transform(data[['Harga', 'Dapur, Kamar Tidur, Hall']])
data_normalisasi = pd.DataFrame(data_normalisasi)
data_normalisasi.rename(columns={0 : 'Harga', 1 : 'Dapur, Kamar Tidur, Hall'}, inplace=True)
data_normalisasi.head(10)
```

Standarisasi

```python
data.drop('Area', axis=1, inplace=True)
data.drop('Kamar Mandi', axis=1, inplace=True)
data.drop('Perabotan', axis=1, inplace=True)
data.drop('Daerah', axis=1, inplace=True)
data.drop('Tempat Parkir', axis=1, inplace=True)
data.drop('Status', axis=1, inplace=True)
data.drop('Transaksi', axis=1, inplace=True)
data.drop('Tipe', axis=1, inplace=True)
data.drop('Ukuran Rumah', axis=1, inplace=True)
```

```python
SScaler = StandardScaler()

data_standarisasi = SScaler.fit_transform(data[['Dapur, Kamar Tidur, Hall', 'Harga']])
print('Dataset sebelum scalling :',
     data[['Dapur, Kamar Tidur, Hall', 'Harga']].head(10),
     f'Nilai standar deviasi : {np.std(data)}',
     sep='\n', end='\n\n')
print('Dataset setelah scalling :',
     data_standarisasi[:10],
     f'Nilai standar deviasi : {np.std(data_standarisasi)}',
     sep='\n')
```

Data cleaning pada data yang mengandung nilai null. 
Menggunakan strategy modus/most frequent

```python
imputer_modus = SimpleImputer(strategy='most_frequent')
```

```python
print(f'Jumlah record yang memiliki nilai null :')
print(data2.isna().sum())

data2[['Halaman', 'Sumber URL']] = imputer_modus.fit_transform(data2[['Halaman', 'Sumber URL']])
print(f'Jumlah record setelah menggunakan fungsi SimpleImputer :')
print(data2.isna().sum())
```
